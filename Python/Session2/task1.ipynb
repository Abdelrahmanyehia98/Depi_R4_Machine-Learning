For Loop : Used when the number of iterations is known

While Loop : Used when the loop depends on a condition, and the number of iterations is not known


break : Stops the loop completely
continue: Skips the current iteration and continues with the next one


```python
def check_number(num1,num2):
 while True:
    num = int(input("Enter a number: "))
    print(num)
    if num == num1 or num == num2:
        print(num, "- exiting the loop")
        break

check_number(5,30)
```

```python
for i in range(1,29):
    print(i,end=" ")
    if i == 5:
        break
    else:
        continue
        print("10")
```
