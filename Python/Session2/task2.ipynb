```python
from countries_details_dat import countries_data,newpar

u_languages = set()

for country in countries_data:
    for Ulang in country['languages']:
        u_languages.add(Ulang)

print("Total number of unique languages:",end = " ")
print(len(u_languages))
```

```python
language_counts = {}

for country in countries_data:
    for lang in country['languages']:
        if lang in language_counts:
            language_counts[lang] += 1
        else:
            language_counts[lang] = 1

most_spoken = sorted(language_counts.items(), key=lambda x: x[1], reverse=True)

print("\nThe 10 most spoken languages are:")
for i in range(10):
    lang_name = most_spoken[i][0]
    print(i+1,"-",end=" " )
    print(lang_name)
```

```python
sorted_countries = sorted(countries_data, key=lambda x: x['population'], reverse=True)

print("The 10 most populated countries in the world:")
for i in range(10):
    name = sorted_countries[i]['name']
    popul = sorted_countries[i]['population']
    print(i+1,"- ",end="")
    print(name,":",end="")
    print(popul)
```
