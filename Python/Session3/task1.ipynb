```python
class my_statistics:
    def __init__(self,data):
        self.data =data

    def count(self):
        return len(self.data)
    
    def sum(self):
        mysum = 0
        for i in self.data:
            mysum = mysum+i
        return mysum
    
    def mean(self):
        return self.sum() / self.count()
    
    def median(self):
        sorted_data = sorted(self.data)
        n = self.count() // 2
        if n%2==0:
            return (sorted_data[n+1]+sorted_data[n-1]) / 2
        else:
            return sorted_data[n]

ages = [31,26,34,37,27,26,32,32,26,27]
calc = my_statistics(ages)
print("count = ",calc.count())
print(calc.sum())
print(calc.mean())
print(calc.median())
    
```
