```python
class voting_system:
    def __init__(self):
        self.candidates = {}  

    def add_candidate(self, name):
        if name in self.candidates:
            print(name,"is already exist")
        else:
            self.candidates[name] = 0
    
    def show_candidates(self):
        if not self.candidates:
            print("not exist")
        else:
            print("Candidates:")
            for name in self.candidates:
                print(f"- {name}")

    def vote(self, name):
        if name in self.candidates:
            self.candidates[name] += 1
        else:
            print(f"error: '{name}' does not exist")

    def get_results(self, name):
        if name in self.candidates:
            print(f"Number of votes for ({name}) are: {self.candidates[name]}")
        else:
            print("No candidate")

    def display_winner(self):
        if not self.candidates:
            print("No candidates to evaluate.")
            return
        winner = max(self.candidates, key=self.candidates.get)
        votes = self.candidates[winner]
        print(f"The winner is {winner} with {votes} votes.")


voting = voting_system()
voting.add_candidate('Abdelrahman')
voting.add_candidate('Yehia')
voting.add_candidate('Elsayed')

voting.show_candidates()
voting.vote('Abdelrahman')
voting.vote('Abdelrahman')
voting.vote('Abdelrahman')
voting.vote('Yehia')
voting.vote('Yehia')
voting.vote('bbdgf')

voting.get_results('Abdelrahman')
voting.get_results('Yehia')
voting.get_results('bbdgf')

voting.display_winner()
```
