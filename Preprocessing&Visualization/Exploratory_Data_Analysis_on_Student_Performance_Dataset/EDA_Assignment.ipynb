```python
import pandas as pd
df = pd.read_csv("student_Dataset.csv", encoding="latin1")
df.head()
```

```python
df.shape
```

```python
print(df.columns)
print(df.info())
```

```python
df['gender'].value_counts()
```

```python
df['gender'] = df['gender'].replace({
    'M': 'Male',
    'male': 'Male',
    'F': 'Female',
    'female': 'Female'
})
df['gender'].value_counts()
```

```python
df.drop(columns=['fNAME', 'lNAME'], inplace=True)
df.columns
```

```python
df['country'].value_counts()
```

```python
df['country'] = df['country'].replace({
    'norway' : 'Norway',
    'Norge' : 'Norway',
    'Rsa' : 'South Africa'
})
```

```python
df['country'].value_counts()
```

```python
df['prevEducation'].value_counts()
```

```python
df['prevEducation'] = df['prevEducation'].replace({
    'Barrrchelors': 'Bachelors',

    'HighSchool': 'High School',

    'Diplomaaa': 'Diploma',
    'DIPLOMA': 'Diploma',
    'diploma': 'Diploma',

})
```

```python
df['prevEducation'].value_counts()
```

```python
df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer

num_imputer = SimpleImputer(strategy='mean')
df['Python'] = num_imputer.fit_transform(df[['Python']])
```

```python
df.isnull().sum()
```

```python
df.to_csv("student_Dataset_cleaned.csv", index=False)
```
